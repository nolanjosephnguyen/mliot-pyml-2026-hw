```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
```

```python
train_path = 'dry_bean_train.csv'
test_path = 'dry_bean_test.csv'

train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_path)

TARGET = 'class'
X_train = train_df.drop(columns=[TARGET])
y_train = train_df[TARGET]

X_test = test_df.drop(columns=[TARGET])
y_test = test_df[TARGET]

print(f'Kích thước tập Train: {X_train.shape}')
print(f'Kích thước tập Test: {X_test.shape}')
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# LOGISTIC REGRESSION

lr_model = LogisticRegression(max_iter=1000, random_state=42)
lr_model.fit(X_train_scaled, y_train)

y_pred_lr = lr_model.predict(X_test_scaled)
```

```python
# KNN

knn_model = KNeighborsClassifier(n_neighbors=5, weights='uniform')
knn_model.fit(X_train_scaled, y_train)

y_pred_knn = knn_model.predict(X_test_scaled)
```

```python
lr_ac = accuracy_score(y_test, y_pred_lr)
knn_ac = accuracy_score(y_test, y_pred_knn)

print(f"BỘ TEST GỒM ({len(y_test)} MẪU)")

print(f"\n1. LOGISTIC REGRESSION ACCURACY: {lr_ac:.4f}")
print(classification_report(y_test, y_pred_lr, digits=4))

print(f"\n2. K-NEAREST NEIGHBORS (K=5) ACCURACY: {knn_ac:.4f}")
print(classification_report(y_test, y_pred_knn, digits=4))
```

Nhận xét: LR và KNN có độ chính xác gần như tương đồng nhau.
